```python
import numpy as np
import pandas as pd
import datawig as dw

df = pd.read_csv("Train_TP2_Datos_2020-2C.csv")

df
```

```python
from operator import itemgetter

nulls = []
nones = []
nats = []
infs = []

for col in df:
    if df[col].isnull().sum() != 0:
        nulls.append((col, df[col].isnull().sum()))
    if df[col].astype('str').str.contains('None').sum() != 0:
        nones.append((col, df[col].astype('str').str.contains('None').sum()))
    if df[col].astype('str').str.contains('NaT').sum() != 0:
        nats.append((col, df[col].astype('str').str.contains('NaT').sum()))
    if ((df[col] == np.inf) | (df[col] == -np.inf)).sum() != 0:
        infs.append((col, ((df[col] == np.inf) | (df[col] == -np.inf)).sum()))
    
total = nulls + nones + nats + infs

total.sort(key=itemgetter(1), reverse=True)
total
```

```python
df = df.drop(['Prod_Category_A', 'Last_Activity', 'Actual_Delivery_Date', 'Currency', 'Price', 'Size', 'Product_Type', 'Product_Category_B', 'Brand', 'Source '], axis=1)
df.columns
```

# Territory

Entreno para Territory, primero reemplazo todos los None que esten asociados a Region Japan por Japan. Esto lo hago ya que todos los Territory que son Japan tienen Region Japan y todos los que tienen Region Japan tienen Territory Japan o None

```python
df = df.set_index('Region')
df.loc['Japan', 'Territory'] = 'Japan'
df = df.reset_index()
df['Territory'].value_counts()
```

```python
df_train = df.replace(to_replace=['None'], value=np.nan)
df_test = df.replace(to_replace=['None'], value=np.nan)

df_train = df_train.dropna(axis=0, subset=['Territory'])
df_test = df_test[df_test['Territory'].isnull()]

df_test
```

```python
#Initialize a SimpleImputer model
imputer = dw.SimpleImputer(
    input_columns=['ID', 'Region', 'Pricing, Delivery_Terms_Quote_Appr',
       'Pricing, Delivery_Terms_Approved', 'Bureaucratic_Code_0_Approval',
       'Bureaucratic_Code_0_Approved', 'Submitted_for_Approval',
       'Bureaucratic_Code', 'Account_Created_Date', 'Billing_Country',
       'Account_Name', 'Opportunity_Name', 'Opportunity_ID',
       'Sales_Contract_No', 'Account_Owner', 'Opportunity_Owner',
       'Account_Type', 'Opportunity_Type', 'Quote_Type', 'Delivery_Terms',
       'Opportunity_Created_Date', 'Quote_Expiry_Date', 'Last_Modified_Date',
       'Last_Modified_By', 'Product_Family', 'Product_Name', 'ASP_Currency',
       'ASP', 'ASP_(converted)_Currency', 'ASP_(converted)',
       'Planned_Delivery_Start_Date', 'Planned_Delivery_End_Date', 'Month',
       'Delivery_Quarter', 'Delivery_Year', 'TRF', 'Total_Amount_Currency',
       'Total_Amount', 'Total_Taxable_Amount_Currency', 'Total_Taxable_Amount',
       'Stage'],
    output_column='Territory',
    output_path = 'Archivos/Imputer_Model/Territory'
)

#Fit an imputer model on the train data
# imputer.fit(train_df=df_train) # Descomentar para entrenar
imputer = dw.SimpleImputer.load('Archivos/Imputer_Model/Territory') # Descomentar para cargar

predictions = imputer.predict(df_test)
```

```python
predictions['Territory_imputed'].value_counts()
```

```python
predictions['Territory'] = predictions['Territory_imputed']
predictions
```

```python
df_trained = pd.concat([df_train, predictions.drop(['Territory_imputed', 'Territory_imputed_proba'], axis=1)])
df_trained
```
